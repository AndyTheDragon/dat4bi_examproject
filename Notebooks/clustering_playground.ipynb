```python
import pandas as pd
from ShowClustering.clustering import (
    load_table,
    prepare_features,
    run_meanshift,
    plot_cluster_sizes,
    plot_centroids_over_pca,
    plot_pca_scatter_2d_named,
    plot_pca_scatter_3d_interactive_named,   # <-- add
    plot_feature_scatter_2d,
    plot_feature_scatter_3d_interactive,     # <-- add
    DEFAULT_NUMERIC_COLS,
    DEFAULT_CATEGORICAL_COLS,
)
```

## 1) Point to your data

```python
DATA_PATH = "data/cleaned_sales_data.csv"   # or 'Book1.xlsx'
SHEET = None               # Only used for Excel
CSV_KWARGS = {}            # e.g., {'sep': ';', 'encoding': 'utf-8-sig'}

df = load_table(DATA_PATH, sheet=SHEET, **CSV_KWARGS)
df.head()
```

## 2) Build features (ignores 'datetime' & 'order_id')

```python
X, num_cols, cat_cols = prepare_features(
    df,
    numeric_cols=DEFAULT_NUMERIC_COLS,
    categorical_cols=DEFAULT_CATEGORICAL_COLS,
    ignore_cols=["datetime", "order_id"],
    force_categorical=["day_of_week", "payment_method", "is_takeaway"],
    auto_detect_categoricals=False,
    force_log1p=["order_total"],
)
print("Numeric used:", num_cols)
print("Categorical used:", cat_cols)
X.head()
```

## 3) Run MeanShift

```python
res = run_meanshift(
    X,
    quantiles=(0.05, 0.10, 0.15, 0.20, 0.25, 0.30),
    pca_variance=0.95,
    min_cluster_size_ratio=0.01,
    centroid_merge_eps=0.75,
)
print("Clusters:", res.n_clusters)
print("Sizes:", res.cluster_sizes)
print("Chosen quantile:", res.selected_quantile, "| bandwidth:", res.selected_bandwidth)
```

## 4) Visualize

```python
# Get encoded matrix before PCA for consistent 2D plots
prep = res.pipeline.named_steps['prep']
X_encoded = prep.transform(X)

plot_cluster_sizes(res.labels)
plot_centroids_over_pca(X_encoded, res.labels, title="Clusters with Centroids (2D PCA)")
```

```python
# Original feature views (pick features meaningful to you)
plot_feature_scatter_2d(X, res.labels, x_col="order_total" if "order_total" in X.columns else X.columns[0], y_col="number_of_drinks" if "number_of_drinks" in X.columns else X.columns[1])
plot_feature_scatter_3d_interactive(                   # <-- switched to interactive
    X, res.labels,
    x_col="order_total" if "order_total" in X.columns else X.columns[0],
    y_col="number_of_drinks" if "number_of_drinks" in X.columns else X.columns[1],
    z_col="number_of_snacks" if "number_of_snacks" in X.columns else X.columns[2],
)
```

```python
# PCA views with axes annotated by top contributing features
plot_pca_scatter_2d_named(res.pipeline, X, res.labels, top_k=3, title_prefix="Clusters (PCA 2D)")
plot_pca_scatter_3d_interactive_named(                 # <-- switched to interactive
    res.pipeline, X, res.labels, top_k=3, title_prefix="Clusters (PCA 3D, interactive)"
)
```
